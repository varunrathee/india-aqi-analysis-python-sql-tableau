```python
import pandas as pd

df = pd.read_csv("/content/city_day.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.dropna(subset=['AQI'])

pollutant_cols = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene']
for col in pollutant_cols:
    df[col] = df.groupby('City')[col].transform(lambda x: x.fillna(x.median()))

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

print(df.groupby('City')['AQI'].mean().sort_values(ascending=False).head(10))
print(df.groupby('City')['AQI'].mean().sort_values().head(10))

df.to_csv("aqi_cleaned.csv", index=False)

import sqlite3
conn = sqlite3.connect("aqi.db")
df.to_sql("aqi", conn, if_exists="replace", index=False)

print(pd.read_sql_query("SELECT City, AVG(AQI) as avg_aqi FROM aqi GROUP BY City ORDER BY avg_aqi DESC LIMIT 10;", conn))
```
